```python
%config InlineBackend.figure_format = 'svg'
import matplotlib as mpl

mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False

t=np.linspace(0,60,40)
f_1=(t-2)/(t+6)+1
f_2=(t-2)/(t+6)+0.5
z=[]
for i in range(40):
    z.append(1)
thresh=np.array(z)
plt.plot(t,f_1,label='Transfer (pre-trained weights)')
plt.plot(t,f_2,label='No Transfer (randomly initialized weights)')
plt.plot(t,z,'--',label='Threshold Performance',color='black',linewidth=1)
plt.xlim([0,60])
plt.ylim([0,2])
plt.yticks(np.arange(0, 2.1, 0.5),fontsize=9)
plt.legend()
plt.xlabel('Training Time (sample complexity)',size=12)
plt.ylabel('Performance',size=12)
plt.savefig('TLLL.png',dpi=400)
```
